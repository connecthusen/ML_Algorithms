```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class SVM:
    def __init__(self, lr=0.001, c=1.0, epochs=1000):
        self.lr = lr
        self.c = c
        self.epochs = epochs
        self.coef_ = None
        self.intercept_ = None

    def fit(self, x_train, y_train):

        X = np.asarray(x_train, dtype=np.float64)
        y = np.asarray(y_train, dtype=np.float64).ravel()

        # Convert labels to {-1, +1}
        y = np.where(y <= 0, -1, 1)

        m, n = X.shape

        self.coef_ = np.zeros(n)
        self.intercept_ = 0.0

        for _ in range(self.epochs):

            # Optional: shuffle every epoch
            idx = np.random.permutation(m)
            X = X[idx]
            y = y[idx]

            for i in range(m):

                margin = y[i] * (np.dot(X[i], self.coef_) + self.intercept_)

                if margin >= 1:
                    dj_dw = self.coef_
                    dj_db = 0.0
                else:
                    dj_dw = self.coef_ - self.c * y[i] * X[i]
                    dj_db = -self.c * y[i]

                self.coef_ -= self.lr * dj_dw
                self.intercept_ -= self.lr * dj_db

        return self

    def decision_function(self, X):
        X = np.asarray(X, dtype=np.float64)
        return np.dot(X, self.coef_) + self.intercept_

    def predict(self, X):
        scores = self.decision_function(X)
        return np.where(scores >= 0, 1, -1)

    def score(self, X, y):
        y = np.where(y <= 0, -1, 1)
        pred = self.predict(X)
        return np.mean(pred == y)

    def plot(self, X, y):

        if X.shape[1] != 2:
            raise ValueError("Plotting is supported only for 2 features.")

        y = np.where(y <= 0, -1, 1)

        plt.figure(figsize=(7,6))

        plt.scatter(
            X[y == -1, 0],
            X[y == -1, 1],
            marker='o',
            label='Class -1'
        )

        plt.scatter(
            X[y == 1, 0],
            X[y == 1, 1],
            marker='^',
            label='Class +1'
        )

        ax = plt.gca()
        xlim = ax.get_xlim()
        ylim = ax.get_ylim()

        xx = np.linspace(xlim[0], xlim[1], 100)
        yy = np.linspace(ylim[0], ylim[1], 100)

        YY, XX = np.meshgrid(yy, xx)

        grid = np.c_[XX.ravel(), YY.ravel()]

        Z = self.decision_function(grid)
        Z = Z.reshape(XX.shape)

        plt.contour(
            XX,
            YY,
            Z,
            levels=[-1, 0, 1],
            linestyles=['--', '-', '--']
        )

        plt.xlabel("Feature 1")
        plt.ylabel("Feature 2")
        plt.legend()
        plt.show()
```

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(
    n_samples=200,
    centers=2,
    random_state=42,
    cluster_std=1.5
)

model = SVM(
    lr=0.001,
    c=1.0,
    epochs=1000
)

model.fit(X, y)

print("Accuracy:", model.score(X, y))

pred = model.predict(X[:50])
print(pred)

model.plot(X, y)
```
