```python
# Decision tree 
import numpy as np
from collections import Counter
```

```python
class create_node:
    def __init__(self,feature = None,threshold = None,feature_type = None,left = None,right = None,value = None)):
        self.feature=feature
        self.threshold=threshold
        self.feature_type=feature_type
        self.right=right
        self.left=left
        self.value=value

    def __repr__(self):
        return self.value
```
